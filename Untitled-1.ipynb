1) 라이브러리 import / 경로 설정

```python
import pandas as pd
import numpy as np
from pathlib import Path

# 표시 옵션
pd.set_option("display.max_columns", 200)
pd.set_option("display.width", 200)
pd.set_option("display.max_rows", 200)

# 파일 경로
file_path = Path(r"C:\Users\EL094\Downloads\rtu_data_full.csv")
output_dir = Path.cwd()

summary_file = output_dir / "eda_summary_by_module.csv"
quality_file = output_dir / "data_quality_report.csv"
suspicious_file = output_dir / "suspicious_rows.csv"

print("Input file:", file_path)
print("Output dir:", output_dir)
```

2) 데이터 로드

```python
df = pd.read_csv(file_path)

# 날짜 컬럼 변환
# 원본 timestamp는 epoch(ms), localtime은 YYYYMMDDHHMMSS 형식이라 각각 분리 변환
if "timestamp" in df.columns:
    df["timestamp"] = pd.to_datetime(df["timestamp"], unit="ms", errors="coerce")

if "localtime" in df.columns:
    df["localtime"] = pd.to_datetime(
        df["localtime"].astype(str),
        format="%Y%m%d%H%M%S",
        errors="coerce"
    )

print("데이터 로드 완료")
print(df.head())
```

3) 기본 정보 확인: shape, 컬럼명, dtype

```python
print("1. 데이터 shape")
print(df.shape)

print("\n2. 컬럼명")
print(df.columns.tolist())

print("\n3. dtype")
print(df.dtypes)
```

4) 분석용 컬럼 정의

```python
module_col = "module(equipment)"
time_col = "timestamp"

numeric_cols = [
    "voltageR", "voltageS", "voltageT",
    "voltageRS", "voltageST", "voltageTR",
    "currentR", "currentS", "currentT",
    "activePower",
    "powerFactorR", "powerFactorS", "powerFactorT",
    "reactivePowerLagging",
    "accumActiveEnergy"
]

# 실제 존재하는 numeric 컬럼만 사용
numeric_cols = [c for c in numeric_cols if c in df.columns]

# 숫자형 변환
for col in numeric_cols:
    df[col] = pd.to_numeric(df[col], errors="coerce")

print("numeric_cols:", numeric_cols)
```

5) 결측치 개수와 비율 확인

```python
missing_report = pd.DataFrame({
    "column": df.columns,
    "missing_count": df.isna().sum().values,
    "missing_ratio_pct": (df.isna().mean().values * 100).round(4),
    "dtype": df.dtypes.astype(str).values
}).sort_values(by="missing_count", ascending=False)

print("4. 결측치 개수 / 비율")
print(missing_report)
```

6) 중복 행 확인

```python
duplicate_count = df.duplicated().sum()
duplicate_rows = df[df.duplicated(keep=False)].copy()

print("5. 중복 행 개수:", duplicate_count)

if duplicate_count > 0:
    print("\n중복 행 샘플:")
    display(duplicate_rows.head(20))
```

7) module별 row count 확인

```python
# module별 적재 건수를 요약해서 설비별 데이터 규모를 비교
module_row_count = (
    df.groupby(module_col, dropna=False)
      .size()
      .reset_index(name="row_count")
      .sort_values("row_count", ascending=False)
)

print("6. module별 row count")
display(module_row_count)
```

8) module별 시작 시간, 종료 시간, 수집 기간 확인

```python
module_time_range = (
    df.groupby(module_col, dropna=False)[time_col]
      .agg(start_time="min", end_time="max")
      .reset_index()
)

module_time_range["collection_duration"] = module_time_range["end_time"] - module_time_range["start_time"]
module_time_range["collection_days"] = module_time_range["collection_duration"].dt.total_seconds() / 86400

print("7. module별 시작/종료/수집기간")
display(module_time_range)
```

9) timestamp 기준 수집 간격이 5초인지 확인

```python
# timestamp 간 차이를 초 단위로 계산해 5초 주기 수집 여부를 점검
df_sorted = df.sort_values([module_col, time_col]).copy()

df_sorted["time_diff_sec"] = (
    df_sorted.groupby(module_col)[time_col]
    .diff()
    .dt.total_seconds()
)

interval_check = (
    df_sorted.groupby(module_col)["time_diff_sec"]
    .agg(
        diff_count="count",
        min_diff_sec="min",
        median_diff_sec="median",
        mean_diff_sec="mean",
        max_diff_sec="max"
    )
    .reset_index()
)

interval_check["is_all_5sec"] = df_sorted.groupby(module_col)["time_diff_sec"].apply(
    lambda s: s.dropna().eq(5).all()
).values

interval_check["non_5sec_count"] = df_sorted.groupby(module_col)["time_diff_sec"].apply(
    lambda s: (~s.dropna().eq(5)).sum()
).values

print("8. module별 수집 간격 5초 여부")
display(interval_check)

print("\n비정상 간격 샘플:")
non_5sec_rows = df_sorted[df_sorted["time_diff_sec"].notna() & (df_sorted["time_diff_sec"] != 5)].copy()
display(non_5sec_rows[[module_col, time_col, "time_diff_sec"]].head(20))
```

10) operation 컬럼 unique 값 확인

```python
print("9. operation unique 값")
if "operation" in df.columns:
    print(df["operation"].dropna().unique().tolist())
    print("\noperation value counts")
    display(df["operation"].value_counts(dropna=False))
else:
    print("'operation' 컬럼이 없습니다.")
```

11) numeric 컬럼별 min, max, mean, std, median 계산

```python
numeric_summary = df[numeric_cols].agg(["min", "max", "mean", "std", "median"]).T.reset_index()
numeric_summary = numeric_summary.rename(columns={"index": "column"})

print("10. numeric 컬럼 요약 통계")
display(numeric_summary)
```

12) module별 activePower 평균/최대/표준편차 계산

```python
active_power_by_module = (
    df.groupby(module_col)["activePower"]
      .agg(activePower_mean="mean",
           activePower_max="max",
           activePower_std="std")
      .reset_index()
)

print("11. module별 activePower 통계")
display(active_power_by_module)
```

13) module별 currentR/S/T 평균과 표준편차 계산

```python
current_stats = (
    df.groupby(module_col)[["currentR", "currentS", "currentT"]]
      .agg(["mean", "std"])
)

current_stats.columns = [f"{col}_{stat}" for col, stat in current_stats.columns]
current_stats = current_stats.reset_index()

print("12. module별 currentR/S/T 평균, 표준편차")
display(current_stats)
```

14) module별 powerFactorR/S/T 평균과 최솟값 계산

```python
pf_stats = (
    df.groupby(module_col)[["powerFactorR", "powerFactorS", "powerFactorT"]]
      .agg(["mean", "min"])
)

pf_stats.columns = [f"{col}_{stat}" for col, stat in pf_stats.columns]
pf_stats = pf_stats.reset_index()

print("13. module별 powerFactorR/S/T 평균, 최솟값")
display(pf_stats)
```

15) module별 reactivePowerLagging 평균/최대 계산

```python
reactive_power_stats = (
    df.groupby(module_col)["reactivePowerLagging"]
      .agg(reactivePowerLagging_mean="mean",
           reactivePowerLagging_max="max")
      .reset_index()
)

print("14. module별 reactivePowerLagging 평균/최대")
display(reactive_power_stats)
```

16) 이상값 후보 자동 탐지

```python
suspicious_parts = []

# 전압 0 이하
voltage_cols = [c for c in ["voltageR", "voltageS", "voltageT", "voltageRS", "voltageST", "voltageTR"] if c in df.columns]
if voltage_cols:
    mask_voltage = df[voltage_cols].le(0).any(axis=1)
    tmp = df.loc[mask_voltage].copy()
    tmp["suspicious_reason"] = "voltage <= 0"
    suspicious_parts.append(tmp)

# 전류 0 이하
current_cols = [c for c in ["currentR", "currentS", "currentT"] if c in df.columns]
if current_cols:
    mask_current = df[current_cols].le(0).any(axis=1)
    tmp = df.loc[mask_current].copy()
    tmp["suspicious_reason"] = "current <= 0"
    suspicious_parts.append(tmp)

# activePower 0 이하
if "activePower" in df.columns:
    mask_active = df["activePower"].le(0)
    tmp = df.loc[mask_active].copy()
    tmp["suspicious_reason"] = "activePower <= 0"
    suspicious_parts.append(tmp)

# powerFactor < 0 or > 100
pf_cols = [c for c in ["powerFactorR", "powerFactorS", "powerFactorT"] if c in df.columns]
if pf_cols:
    mask_pf = (df[pf_cols].lt(0) | df[pf_cols].gt(100)).any(axis=1)
    tmp = df.loc[mask_pf].copy()
    tmp["suspicious_reason"] = "powerFactor out of range (<0 or >100)"
    suspicious_parts.append(tmp)

# accumActiveEnergy 감소
if "accumActiveEnergy" in df.columns:
    df_energy = df.sort_values([module_col, time_col]).copy()
    df_energy["accum_diff"] = df_energy.groupby(module_col)["accumActiveEnergy"].diff()
    mask_energy_drop = df_energy["accum_diff"] < 0
    tmp = df_energy.loc[mask_energy_drop].copy()
    tmp["suspicious_reason"] = "accumActiveEnergy decreased"
    suspicious_parts.append(tmp)

# suspicious rows 통합
if suspicious_parts:
    # 여러 규칙에서 동시에 걸린 행을 합치고 중복 제거
    suspicious_rows = pd.concat(suspicious_parts, axis=0, ignore_index=True)
    suspicious_rows = suspicious_rows.drop_duplicates()
else:
    suspicious_rows = pd.DataFrame()

print("15. 이상값 후보 건수:", len(suspicious_rows))
display(suspicious_rows.head(20))
```

17) module별 종합 요약 테이블 생성

```python
eda_summary_by_module = module_row_count.merge(module_time_range, on=module_col, how="left")
eda_summary_by_module = eda_summary_by_module.merge(interval_check, on=module_col, how="left")
eda_summary_by_module = eda_summary_by_module.merge(active_power_by_module, on=module_col, how="left")
eda_summary_by_module = eda_summary_by_module.merge(current_stats, on=module_col, how="left")
eda_summary_by_module = eda_summary_by_module.merge(pf_stats, on=module_col, how="left")
eda_summary_by_module = eda_summary_by_module.merge(reactive_power_stats, on=module_col, how="left")

print("16. eda_summary_by_module")
display(eda_summary_by_module)
```

18) data_quality_report 생성

```python
quality_items = []

# 전체 데이터 기본 품질
quality_items.append({
    "category": "dataset",
    "item": "row_count",
    "value": len(df),
    "detail": ""
})
quality_items.append({
    "category": "dataset",
    "item": "column_count",
    "value": df.shape[1],
    "detail": ""
})
quality_items.append({
    "category": "dataset",
    "item": "duplicate_row_count",
    "value": int(duplicate_count),
    "detail": ""
})
quality_items.append({
    "category": "dataset",
    "item": "duplicate_row_ratio_pct",
    "value": round(duplicate_count / len(df) * 100, 4) if len(df) > 0 else np.nan,
    "detail": ""
})

# 컬럼별 결측치
for _, row in missing_report.iterrows():
    quality_items.append({
        "category": "missing",
        "item": row["column"],
        "value": row["missing_count"],
        "detail": f'missing_ratio_pct={row["missing_ratio_pct"]}, dtype={row["dtype"]}'
    })

# 수집 간격 이슈
for _, row in interval_check.iterrows():
    quality_items.append({
        "category": "interval",
        "item": row[module_col],
        "value": row["non_5sec_count"],
        "detail": f'is_all_5sec={row["is_all_5sec"]}, min={row["min_diff_sec"]}, median={row["median_diff_sec"]}, max={row["max_diff_sec"]}'
    })

# 이상값 개수
if not suspicious_rows.empty:
    suspicious_reason_counts = suspicious_rows["suspicious_reason"].value_counts(dropna=False)
    for reason, cnt in suspicious_reason_counts.items():
        quality_items.append({
            "category": "suspicious",
            "item": reason,
            "value": int(cnt),
            "detail": ""
        })
else:
    quality_items.append({
        "category": "suspicious",
        "item": "no_suspicious_rows",
        "value": 0,
        "detail": ""
    })

data_quality_report = pd.DataFrame(quality_items)

print("17. data_quality_report")
display(data_quality_report.head(50))
```

21) matplotlib, seaborn 시각화

```python
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_theme(style="whitegrid", context="talk")
plt.rcParams["figure.figsize"] = (14, 6)
plt.rcParams["axes.unicode_minus"] = False

# 시각화는 전체 데이터를 그대로 쓰기 무거워서 최대 20만건만 샘플링
viz_sample_size = min(200000, len(df))
viz_df = df.sample(n=viz_sample_size, random_state=42).copy() if len(df) > viz_sample_size else df.copy()

print(f"시각화용 샘플 크기: {len(viz_df):,}")
```

```python
# 1. 모듈별 row 수
plt.figure(figsize=(14, 6))
sns.barplot(data=module_row_count, x=module_col, y="row_count", hue=module_col, dodge=False, palette="Blues_d", legend=False)
plt.title("Module별 데이터 건수")
plt.xlabel("Module")
plt.ylabel("Row Count")
plt.xticks(rotation=45, ha="right")
plt.tight_layout()
plt.show()

# 2. 모듈별 수집 기간
plt.figure(figsize=(14, 6))
sns.barplot(data=module_time_range.sort_values("collection_days", ascending=False), x=module_col, y="collection_days", hue=module_col, dodge=False, palette="Greens_d", legend=False)
plt.title("Module별 수집 기간(일)")
plt.xlabel("Module")
plt.ylabel("Collection Days")
plt.xticks(rotation=45, ha="right")
plt.tight_layout()
plt.show()

# 3. 모듈별 5초 간격이 아닌 건수
plt.figure(figsize=(14, 6))
sns.barplot(data=interval_check.sort_values("non_5sec_count", ascending=False), x=module_col, y="non_5sec_count", hue=module_col, dodge=False, palette="OrRd", legend=False)
plt.title("Module별 5초 간격이 아닌 건수")
plt.xlabel("Module")
plt.ylabel("Non-5sec Count")
plt.xticks(rotation=45, ha="right")
plt.tight_layout()
plt.show()
```

```python
# 4. 주요 수치형 변수 분포
dist_cols = [c for c in ["voltageR", "currentR", "activePower", "reactivePowerLagging", "accumActiveEnergy"] if c in numeric_cols]

if dist_cols:
    fig, axes = plt.subplots(len(dist_cols), 1, figsize=(14, 4 * len(dist_cols)))
    if len(dist_cols) == 1:
        axes = [axes]

    for ax, col in zip(axes, dist_cols):
        sns.histplot(viz_df[col].dropna(), bins=50, kde=True, ax=ax, color="#3b82f6")
        ax.set_title(f"{col} 분포")
        ax.set_xlabel(col)
        ax.set_ylabel("Count")

    plt.tight_layout()
    plt.show()
```

```python
# 5. 수치형 변수 상관관계 히트맵
corr_sample_cols = [c for c in numeric_cols if c in viz_df.columns]
corr_matrix = viz_df[corr_sample_cols].corr(numeric_only=True)

plt.figure(figsize=(14, 10))
sns.heatmap(corr_matrix, cmap="coolwarm", center=0, annot=False, square=True)
plt.title("수치형 변수 상관관계 히트맵")
plt.tight_layout()
plt.show()
```

```python
corr_sample_cols = [c for c in numeric_cols if c in viz_df.columns]
corr_matrix = viz_df[corr_sample_cols].corr(numeric_only=True)

plt.figure(figsize=(14, 10))

sns.heatmap(
    corr_matrix,
    cmap="coolwarm",
    center=0,
    annot=True,      # 셀 안에 수치 표시
    fmt=".2f",       # 소수점 둘째 자리까지 표시
    square=True,
    linewidths=0.5,  # 셀 경계선
    cbar_kws={"label": "Correlation"}
)

plt.title("수치형 변수 상관관계 히트맵")
plt.tight_layout()
plt.show()
```

```python
# 6. 대표 모듈 시계열 추이
target_module = module_row_count.sort_values("row_count", ascending=False).iloc[0][module_col]
time_series_cols = [c for c in ["activePower", "voltageR", "currentR"] if c in df.columns]

ts_df = df.loc[df[module_col] == target_module, [module_col, time_col] + time_series_cols].copy()
ts_df = ts_df.sort_values(time_col)
ts_df = ts_df.iloc[::max(len(ts_df) // 2000, 1), :]

fig, axes = plt.subplots(len(time_series_cols), 1, figsize=(15, 4 * len(time_series_cols)), sharex=True)
if len(time_series_cols) == 1:
    axes = [axes]

for ax, col in zip(axes, time_series_cols):
    sns.lineplot(data=ts_df, x=time_col, y=col, ax=ax, color="#2563eb")
    ax.set_title(f"{target_module} - {col} 시계열 추이")
    ax.set_xlabel("Timestamp")
    ax.set_ylabel(col)

plt.tight_layout()
plt.show()
```
